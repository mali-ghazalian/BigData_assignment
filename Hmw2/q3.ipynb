```python
import pandas as pd
import numpy as np
from itertools import combinations
```

```python
# Reading line by line into a list
with open('browsing.txt', 'r') as file:
    data = file.readlines()
```

```python
baskets = []
for line in data:
    baskets.append([product for product in line.split()])
```

```python
items = []
for basket in baskets:
    for item in basket:
        if item not in items:
            items.append(item)
print(f"Number of items:{len(items)}")
```

<strong style="font-size:38pt;"> A-Priori: </strong>

---
- # First Pass --->
---

```python
# Count singletones
singleton_counts = [0 for _ in items]
for basket in baskets:
    for item in basket:
        item_idx = items.index(item)
        singleton_counts[item_idx] += 1
print(len(singleton_counts))
```

```python
frequent_singletons = []
for i, count in enumerate(singleton_counts):
    if count >= 100:
        frequent_singletons.append(i)
```

```python
doubletons_condidates = list(combinations(frequent_singletons, 2))
```

```python
print(f"Number of doubleton candidates: {len(doubletons_condidates)}")
print(f"Number of items: {len(singleton_counts)}")
print(f"Number of Frequent items: {len(frequent_singletons)}")
print(frequent_singletons)
```


---
- # Second pass --->
---

```python
doubletons_counts = {}
for basket in baskets:
    valid_items = [item for item in frequent_singletons if items[item] in basket]
    for i in valid_items:
        for j in valid_items:
            if i < j:
                pair = tuple(sorted((i, j)))
                if pair in doubletons_counts:
                    doubletons_counts[pair] += 1
                else:
                    doubletons_counts[pair] = 1
```

```python
frequent_doubletons = {pair for pair, count in doubletons_counts.items() if count >= 100}
```

```python
print(f"Number of frequent doubletones: {len(frequent_doubletons)}")
```

### Compute Confidence for rules

```python
ass_rules = {}
for i in frequent_doubletons:
    ass_rules[(items[i[0]], items[i[1]])] = float("{:.5f}".format(doubletons_counts[i] / singleton_counts[i[0]]))
    ass_rules[(items[i[1]], items[i[0]])] = float("{:.5f}".format(doubletons_counts[i] / singleton_counts[i[1]]))
print(ass_rules)
```

```python
sorted_rules = sorted(ass_rules.items(), key=lambda x: x[1], reverse=True)
five_top_double_rules = sorted_rules[:5]
print(five_top_double_rules)
```


---
- # Third pass --->
---

```python
triplet_counts = {}

for basket in baskets:
    valid_triplets = set()
    valid_pairs = [pair for pair in frequent_doubletons if items[pair[0]] in basket and items[pair[1]] in basket]
    
    for i in valid_pairs:
        for j in valid_pairs[valid_pairs.index(i) + 1:]:

            candidate_items = set(i).union(set(j))
            
            if len(candidate_items) == 3:
                
                double_subsets = combinations(candidate_items, 2)
                
                if all(ds in frequent_doubletons for ds in double_subsets):
                    triplet = tuple(sorted(candidate_items))
                    valid_triplets.add(triplet)
                    
    for tri in valid_triplets:
        triplet_counts[tri] = triplet_counts.get(tri, 0) + 1
```

```python
frequent_triplets = []
for i in triplet_counts:
    if triplet_counts[i] >= 100:
        frequent_triplets.append(i)

frequent_triplets = sorted(frequent_triplets)
print(frequent_triplets)
print(f"Number of frequent Triplets: {len(frequent_triplets)}")
```

### Compute Confidence for rules

```python
ass_triplet_rules = {}
for i in frequent_triplets:
    ass_triplet_rules[((items[i[0]], items[i[1]]), items[i[2]])] = triplet_counts[i] / doubletons_counts[(i[0], i[1])]
    ass_triplet_rules[((items[i[0]], items[i[2]]), items[i[1]])] = triplet_counts[i] / doubletons_counts[(i[0], i[2])]
    ass_triplet_rules[((items[i[1]], items[i[2]]), items[i[0]])] = triplet_counts[i] / doubletons_counts[(i[1], i[2])]

print(ass_triplet_rules)
```

```python
sorted_triplet_rules = sorted(ass_triplet_rules.items(), key=lambda x: x[1], reverse=True)
five_top_triplet_rules = sorted_triplet_rules[:5]
print(sorted(five_top_triplet_rules))
```
