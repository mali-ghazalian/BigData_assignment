```python
import pandas as pd
import numpy as np
from itertools import combinations
import random
```

```python
# Reading line by line into a list
with open('../q3/browsing.txt', 'r') as file:
    data = file.readlines()
```

```python
baskets = []
for line in data:
    baskets.append([product for product in line.split()])
```

```python
items = []
for basket in baskets:
    for item in basket:
        if item not in items:
            items.append(item)
```

```python
print(f"Number of items:{len(items)}")
print(f"Number of baskets:{len(baskets)}")
```


---

- # First pass -> 
---

```python
num_buckets = 50000
# Pass 1: Count single items and hash pairs
buckets = [0 for _ in range(num_buckets)]
```

```python
singleton_counts = [0 for _ in items]
for basket in baskets:
    basket_items = []
    basket_set = set(basket)
    for item in basket_set:
        
        idx = items.index(item)
        singleton_counts[idx] += 1
        basket_items.append(idx)
        
    pair_combination = combinations(basket_items, 2)
    
    for pair in pair_combination:
        
        bucket_id = hash(pair) % num_buckets
        buckets[bucket_id] += 1
```

```python
print("Singleton Counts:", singleton_counts)
print("Buckets:", buckets)
```

```python
# Bit map
for bucket_id, count in enumerate(buckets):
    if count > 100:
        buckets[bucket_id] = 1
    else: buckets[bucket_id] = 0
```

```python
print(buckets)
```

```python
frequent_items = []
for i, count in enumerate(singleton_counts):
    if count >= 100:
        frequent_items.append(i)
```

```python
print(f"Number of frequent items: {frequent_items}")
print(f"Number of frequent items: {len(frequent_items)}")
```

```python
doubletons_condidates = set()
valid_pair_combination = combinations(frequent_items, 2)
for pair in valid_pair_combination:
    bucket_id = hash(pair) % num_buckets
    if buckets[bucket_id] == 1:
        doubletons_condidates.add(pair)
```

```python
print(f"Number of doubletons condidates : {len(doubletons_condidates)}")
```

    
---
- # Second pass -> 
---

```python
doubletons_counts = {}
for basket in baskets:
    valid_pairs = [pair for pair in doubletons_condidates if items[pair[0]] in basket and items[pair[1]] in basket]
    for pair in valid_pairs:
        if pair in doubletons_counts:
            doubletons_counts[pair] += 1
        else:
            doubletons_counts[pair] = 1
```

```python
print(doubletons_counts)
```

```python
frequent_doubletons = {pair for pair, count in doubletons_counts.items() if count >= 100}
```

```python
print("Number Frequent doubletons:", len(frequent_doubletons))
print("Frequent doubletons:", frequent_doubletons)
```

```python
for i in frequent_doubletons:
    for j in frequent_doubletons:
        print(i,":",j)
```

```python
triplet_condidates = set()
for i in frequent_doubletons:
    for j in frequent_doubletons:
        if i < j:
            candidate_items = set(i).union(set(j))
            if len(candidate_items) == 3:
                
                double_subsets = combinations(candidate_items, 2)
                    
                if all(ds in frequent_doubletons for ds in double_subsets):
                    triplet = tuple(sorted(candidate_items))
                    triplet_condidates.add(triplet)
```

### Compute Confidence for rules

```python
ass_rules = {}
for i in frequent_doubletons:
    ass_rules[(i[0], i[1])] = float("{:.5f}".format(doubletons_counts[i] / singleton_counts[i[0]]))
    ass_rules[(i[1], i[0])] = float("{:.5f}".format(doubletons_counts[i] / singleton_counts[i[1]]))

sorted_rules = sorted(ass_rules.items(), key=lambda x: x[1], reverse=True)
top_doubleton_rules = [rule for rule in sorted_rules if rule[1]>0.5]
print(f"Top rules with confidence above 0.5:\n {top_doubleton_rules}")
print(f"\n Number of top rules: {len(top_doubleton_rules)}")
```

- # Third pass --->
---

```python
triplet_counts = {}

for basket in baskets:
    valid_triplets = set()
    valid_pairs = [pair for pair in frequent_doubletons if items[pair[0]] in basket and items[pair[1]] in basket]
    
    for i in valid_pairs:
        for j in valid_pairs[valid_pairs.index(i) + 1:]:

            candidate_items = set(i).union(set(j))
            
            if len(candidate_items) == 3:
                
                double_subsets = combinations(candidate_items, 2)
                
                if all(ds in frequent_doubletons for ds in double_subsets):
                    triplet = tuple(sorted(candidate_items))
                    valid_triplets.add(triplet)
                    
    for tri in valid_triplets:
        triplet_counts[tri] = triplet_counts.get(tri, 0) + 1
```

```python
print(triplet_counts)
```

```python
frequent_triplets = []
for i in triplet_counts:
    if triplet_counts[i] >= 100:
        frequent_triplets.append(i)

frequent_triplets = sorted(frequent_triplets)
print(frequent_triplets)
print(f"Number of frequent Triplets: {len(frequent_triplets)}")
```

### Compute Confidence for rules

```python
ass_triplet_rules = {}
for i in frequent_triplets:
    ass_triplet_rules[(i[0], i[1], i[2])] = triplet_counts[i] / doubletons_counts[(i[0], i[1])]
    ass_triplet_rules[(i[0], i[2], i[1])] = triplet_counts[i] / doubletons_counts[(i[0], i[2])]
    ass_triplet_rules[(i[1], i[2], i[0])] = triplet_counts[i] / doubletons_counts[(i[1], i[2])]

sorted_triplet_rules = sorted(ass_triplet_rules.items(), key=lambda x: x[1], reverse=True)
top_triplet_rules = [rule for rule in sorted_triplet_rules if rule[1]>0.5]
print(f"Top rules with confidence above 0.5:\n {top_triplet_rules}")
print(f"\n Number of top rules: {len(top_triplet_rules)}")
```
