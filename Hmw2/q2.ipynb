```python
import chardet
import re
import time
import hashlib
import statistics
import mmh3
import math
import matplotlib.pyplot as plt
```

```python
# Prepare text document file
with open("Shakespeare.txt", 'rb') as file:
    raw = file.read()
    
encoding = chardet.detect(raw)['encoding']
text = raw.decode(encoding)
text = text.lower()
text = re.sub(r'[^\w\s]', '', text)
text = re.sub(r'\d+', '', text)

words = text.split()

with open("cleaned_text.txt", 'w') as cleaned_text:
    cleaned_text.write(text)
```

```python
# Stream function
def stream_words(words, delay=1):
    for word in words:
        yield word
        time.sleep(delay)
```

```python
def trailing_zeros(x):
    if x == 0:
        return 1
    
    s = str(x)
    return len(s)-len(s.rstrip('0'))
```

```python
# hash functions
def hash_function(seed, word, n_bin):
    hash_value = mmh3.hash(word, seed)
    # word_bin = bin(int(hash_value, 16))[2:].zfill(n_bin)
    hash_bin = bin(hash_value & ((1 << n_bin) - 1))[2:].zfill(n_bin)
    
    return hash_bin
```

```python
# Approximation based mean
def approx_mean(result_tail):
    tails = []
    for i in result_tail:
        tails.append(2**(max(i)))
    return statistics.mean(tails)
```

```python
# Approximation based median
def approx_median(result_tail):
    tails = []
    for i in result_tail:
        tails.append(2**(max(i)))
    return statistics.median(tails)
    
```

```python
# approximation based both
# into n baches
def approx_combine(result_tail, n):
    medians = []
    for i in range(n):
        group = []
        for j in result_tail[i:i+n]:
            group.append(2**max(j))
        medians.append(statistics.median(group))
    return statistics.mean(medians)
```

```python
# Get stream data
hash_num = 35
result_tail = [[] for i in range(hash_num)]
i = 0
for word in stream_words(words, 0.1):
    for seed in range(hash_num):
        result = hash_function(seed, word, 24)
        result_tail[seed].append(trailing_zeros(result))
    
    print(f"Stream {i+1}")
    print(f" Mean approx: {approx_mean(result_tail)} \n Median approx: {approx_median(result_tail)} \n Combine Approx: {approx_combine(result_tail, 10)} \n")
    

    i += 1
    if i > 100:
        break
    
```

```python
num_hash_funcs = [10, 30, 200]
bit_len = [16, 24, 32]
result = {}
mean_=[]
median_=[]
both_=[]
for k in num_hash_funcs:
    mean_k = []
    median_k = []
    both_k = []
    for n in bit_len:
        result_tail = [[] for i in range(k)]
        i = 0
        for word in stream_words(words, 0.1):
            for seed in range(k):
                result = hash_function(seed, word, n)
                result_tail[seed].append(trailing_zeros(result))
            mean_k.append(approx_mean(result_tail))
            median_k.append(approx_median(result_tail))
            both_k.append(approx_combine(result_tail, 10))
    
            i += 1
            if i > 50:
                break
    mean_.append(mean_k)
    median_.append(median_k)
    both_.append(both_k)
```
