```python
import string
import chardet
import re
```

```python
file_list = [f"{i:02d}.txt" for i in range(1,6)]
for file_name in file_list:
    
    with open(f"data/{file_name}", 'rb') as file:
        raw = file.read()
    encoding = chardet.detect(raw)['encoding']
    
    text = raw.decode(encoding)
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\¶|§', '', text)
    text = re.sub(r'[·]+', ' ', text)
    with open(f"data/cleaned_{file_name}", 'w') as cleaned_file:
        cleaned_file.write(text)
```

```python
# Remove stop words
stop_words = {"i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"}
single_letters = set("abcdefghijklmnopqrstuvwxyz")

cleaned_file_list = [f"cleaned_{i:02d}.txt" for i in range(1,6)]
for file_name in cleaned_file_list:
    with open(f"data/{file_name}") as file:
        text = file.read()
    words = text.split()
    cleaned_words = [word for word in words if word not in stop_words and word not in single_letters]
    cleaned_text = ' '.join(cleaned_words)
    with open(f"data/stopwords_cleaned_{file_name}", 'w') as cleaned_file:
        cleaned_file.write(cleaned_text)
```
