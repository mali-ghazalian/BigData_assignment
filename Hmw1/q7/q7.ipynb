# Question 7: LSH in ANN

```python
import random
import numpy as np
import hashlib
```

```python
def generate_dataset(n, dimension, min_value=0, max_value=100):
    
    dataset = np.random.uniform(min_value, max_value, size=(n, dimension))
    return dataset
```

```python
def generate_hamming_dataset(n, dimension):
    dataset = np.random.randint(2, size=(n, dimension))
    return dataset
```

```python
n_points = 1000
dimension = 32 
hamming_dataset = generate_hamming_dataset(n_points, dimension)
print(hamming_dataset)
```

```python
L = 10 
z_index = np.random.randint(n_points)
z = hamming_dataset[z_index]
print("query point:", z)
```

```python
def hash_function_generator(num_hf):
    hash_functions = [hashlib.sha256(str(i).encode()).hexdigest() for i in range(num_hf)]
    return hash_functions
```

```python
num_hf = 10
hashed_points = {}
hash_functions = hash_function_generator(num_hf)
for i, data_point in enumerate(hamming_dataset):
    hashed_points[i] = []
    for hf in hash_functions:
        hashed_val = int(hf, 16) % sum(bit == 1 for bit in data_point)
        hashed_points[i].append(hashed_val)
```

```python
hashed_points
```

```python
candidate_points = []
z_sig = hashed_points[z_index]
for data_point, sig in hashed_points.items():
    if sig == z_sig:
        candidate_points.append(data_point)
```

```python
len(candidate_points)
```

```python
def hamming_distance(vec1, vec2):
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    return np.sum(vec1 != vec2)
```

```python
random_indxs = random.sample(range(len(candidate_points)), 30)
distances = {}
for i in random_indxs:
    distances[i] = hamming_distance(hamming_dataset[i], hamming_dataset[z_index])
```

```python
sorted_items = sorted(distances.items(), key=lambda x: x[1])
min_values = sorted_items[:3]

print("Three min values:", min_values)
```
